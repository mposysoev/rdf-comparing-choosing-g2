```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams.update({'font.size': 18})
```

```python
loss1 = pd.read_fwf("short-it-progon1.dat", header=None)

for i in range(len(loss1[0])):
    loss1[0].iloc[i] = float(loss1[0].iloc[i].split("=")[1])
```

```python
plt.figure(figsize=(15,6))
plt.plot(loss1, label="Loss Function")
plt.xlabel("Iteration number")
plt.ylabel("Loss Value")
plt.plot(loss1.rolling(window=15).mean(), label="Average 15")
plt.plot(loss1.cummin(), label="Cumulative Minimum", color='g')
plt.plot(loss1[::-1].cummax()[::-1], label="Reversed Cumulative Maximum", color='r')
plt.ylim(0.25)
plt.legend()
plt.title("Поведение Loss-фукнции при количестве МК-шагов 300к")
```
