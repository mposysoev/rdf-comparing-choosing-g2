# Выбор G2 для обучения

```python
model_rdf100 = np.genfromtxt("RDFNN-100CH3OH-CG-iter-96.dat").T
reference_rdf100 = np.genfromtxt("100CH3OH-CG.rdf").T

model_rdf60 = np.genfromtxt("RDFNN-60CH3OH-CG-iter-96.dat").T
reference_rdf60 = np.genfromtxt("60CH3OH-CG.rdf").T

model_rdf40 = np.genfromtxt("RDFNN-40CH3OH-CG-iter-96.dat").T
reference_rdf40 = np.genfromtxt("40CH3OH-CG.rdf").T

model_rdf10 = np.genfromtxt("RDFNN-10CH3OH-CG-iter-96.dat").T
reference_rdf10 = np.genfromtxt("10CH3OH-CG.rdf").T
```

```python
plt.figure(figsize=(20,8))

plt.plot(reference_rdf100[0,:], reference_rdf100[1,:], label="ref 100", c='darkblue')
plt.plot(model_rdf100[0,:], model_rdf100[1,:], label="model 100", c='blue')
plt.plot(model_rdf100[0,:], reference_rdf100[1,:]-model_rdf100[1,:], label="delta 100", c='deepskyblue')

plt.plot(reference_rdf10[0,:], reference_rdf10[1,:], label="ref 10", c='darkgreen')
plt.plot(model_rdf10[0,:], model_rdf10[1,:], label="model 10", c='limegreen')
plt.plot(model_rdf10[0,:], reference_rdf10[1,:]-model_rdf10[1,:], label="delta 10", c='springgreen')


plt.legend()
plt.title("100 and 10 concentrations")
plt.xlabel("r, Å")
plt.ylabel("RDF(r)")
plt.xlim(2.5, 5.5)
```

```python
plt.figure(figsize=(20, 8))


delta100 = reference_rdf100[1, :] - model_rdf100[1, :]
delta60 = reference_rdf60[1, :] - model_rdf60[1, :]
delta40 = reference_rdf40[1, :] - model_rdf40[1, :]
delta10 = reference_rdf10[1, :] - model_rdf10[1, :]

plt.plot(model_rdf100[0, :], delta100, label="delta 100", c="deepskyblue")
plt.plot(model_rdf60[0, :], delta60, label="delta 60", c="fuchsia")
plt.plot(model_rdf40[0, :], delta40, label="delta 40", c="mediumslateblue")
plt.plot(model_rdf10[0, :], delta10, label="delta 10", c="springgreen")

# plt.plot(model_rdf100[0, :], (np.abs(delta100) + np.abs(delta60) + np.abs(delta40) + np.abs(delta10)), label="Squared Sum", c='black')

plt.legend()
plt.title("Delta = Reference - Model")
plt.xlabel("r, Å")
# plt.ylabel("RDF(r)")
```

```python
plt.figure(figsize=(20, 8))


delta100 = reference_rdf100[1, :] - model_rdf100[1, :]
delta60 = reference_rdf60[1, :] - model_rdf60[1, :]
delta40 = reference_rdf40[1, :] - model_rdf40[1, :]
delta10 = reference_rdf10[1, :] - model_rdf10[1, :]

plt.plot(model_rdf100[0, :], delta100, label="delta 100", c="deepskyblue")
plt.plot(model_rdf60[0, :], delta60, label="delta 60", c="fuchsia")
plt.plot(model_rdf40[0, :], delta40, label="delta 40", c="mediumslateblue")
plt.plot(model_rdf10[0, :], delta10, label="delta 10", c="springgreen")

plt.plot(model_rdf100[0, :], (np.abs(delta100) + np.abs(delta60) + np.abs(delta40) + np.abs(delta10)), label="Abs error Sum", c='black')

plt.legend()
plt.title("Delta = Reference - Model")
plt.xlabel("r, Å")
# plt.ylabel("RDF(r)")
```

```python
plt.figure(figsize=(20, 8))


delta100 = reference_rdf100[1, :] - model_rdf100[1, :]
delta60 = reference_rdf60[1, :] - model_rdf60[1, :]
delta40 = reference_rdf40[1, :] - model_rdf40[1, :]
delta10 = reference_rdf10[1, :] - model_rdf10[1, :]

plt.plot(model_rdf100[0, :], (np.abs(delta100) + np.abs(delta60) + np.abs(delta40) + np.abs(delta10)), label="Abs error Sum", c='black')
plt.plot(model_rdf100[0, :], (np.power(delta100, 2) + np.power(delta60, 2) + np.power(delta40, 2) + np.power(delta10, 2)), label="Squared Sum", c='grey')

plt.legend()
plt.title("Delta = Reference - Model")
plt.xlabel("r, Å")
# plt.ylabel("RDF(r)")
```

# Detecting noise for different amount of steps

```python
rdf_ref = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-100M.dat").T # Здесь должно быть 100 млн

rdf_5m = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-5M.dat").T
rdf_2m = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-2M.dat").T
rdf_1m = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-1M.dat").T
rdf_900k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-900k.dat").T
rdf_800k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-800k.dat").T
rdf_700k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-700k.dat").T
rdf_600k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-600k.dat").T
rdf_500k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-500k.dat").T
rdf_400k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-400k.dat").T
rdf_300k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-300k.dat").T
rdf_200k = np.genfromtxt("rdf-noise/RDFNN-100CH3OH-CG-200k.dat").T
```

```python
all_rdf = [rdf_5m,rdf_2m,rdf_1m,rdf_900k,rdf_800k,rdf_700k,rdf_600k,rdf_500k,rdf_400k,rdf_300k,rdf_200k]
all_rdf_names = ["5M","2M","1M","900k","800k","700k","600k","500k","400k","300k","200k"]
all_steps = [5000, 2000, 1000, 900, 800, 700, 600, 500, 400, 300, 200]
```

```python
plt.figure(figsize=(20, 8))
plt.plot(rdf_ref[0, :], rdf_5m[1, :], lw=6, color="tab:red")
plt.plot(rdf_ref[0, :], rdf_5m[1, :], label="RDF 5M MK steps", lw=4, color="tab:orange")
plt.plot(
    rdf_ref[0, :],
    rdf_ref[1, :],
    label="Reference (100M)",
    lw=1,
    c="black")
plt.plot(rdf_ref[0, :], rdf_ref[1, :] - rdf_1m[1, :], label="Delta", c="tab:green")
plt.legend()

plt.xlabel("r, Å")
plt.ylabel("RDF(r)")
plt.title("Best")
```

```python
plt.figure(figsize=(20, 8))
plt.plot(rdf_ref[0, :], rdf_200k[1, :], lw=6, color="tab:red")
plt.plot(
    rdf_ref[0, :], rdf_200k[1, :], label="RDF 200k MK steps", lw=4, color="tab:orange"
)

plt.plot(
    rdf_ref[0, :],
    rdf_ref[1, :],
    label="Reference (100M)",
    lw=1,
    c="black",
    # linestyle=(5, (10, 0.5)),
)
plt.plot(rdf_ref[0, :], rdf_ref[1, :] - rdf_200k[1, :], label="Delta", c="tab:green")
plt.legend()

plt.xlabel("r, Å")
plt.ylabel("RDF(r)")
plt.title("Worst")
```

```python
plt.figure(figsize=(20, 8))

delta200k = rdf_ref[1, :] - rdf_200k[1, :]
delta5m = rdf_ref[1, :] - rdf_5m[1, :]

plt.plot(rdf_ref[0, :], delta200k, label=f"Worst L = {round(np.sum(delta200k**2), 6)}")
plt.plot(rdf_ref[0, :], delta5m, label=f"Best L = {round(np.sum(delta5m**2), 6)}")
plt.legend()

plt.xlabel("r, Å")
plt.ylabel("Delta")
plt.title("Worst vs Best delta")
```

```python
plt.figure(figsize=(20,8))
for i, rdf in enumerate(all_rdf): 
    delta = rdf_ref[1,:] - rdf[1,:]
    loss = np.sum(delta**2)
    plt.plot(rdf_ref[0,:], rdf_ref[1,:] - rdf[1,:], label=f"{all_rdf_names[i]} L={round(loss, 5)}")
plt.legend(loc=2, title="Ошибка из-за статистики")
plt.xlabel("r, Å")
plt.ylabel("Delta")
```

```python
plt.figure(figsize=(20,8))

delta = reference_rdf100[1,:] - rdf_ref[1,:]
loss = np.sum(delta**2)
plt.plot(reference_rdf100[0,:], reference_rdf100[1,:] - rdf_ref[1,:], c='red', lw=13)
plt.plot(reference_rdf100[0,:], reference_rdf100[1,:] - rdf_ref[1,:], label=f"100M L={round(loss, 5)}", c='lightpink', lw=12)

for i, rdf in enumerate(all_rdf): 
    delta = reference_rdf100[1,:] - rdf[1,:]
    loss = np.sum(delta**2)
    plt.plot(reference_rdf100[0,:], reference_rdf100[1,:] - rdf[1,:], label=f"{all_rdf_names[i]} L={round(loss, 5)}")
plt.legend(loc=2, title="Итоговый Loss")
plt.xlabel("r, Å")
plt.ylabel("Delta")
```

```python
plt.figure(figsize=(20,8))

delta = reference_rdf100[1,:] - rdf_ref[1,:]
loss = np.sum(delta**2)
# plt.plot(reference_rdf100[0,:], reference_rdf100[1,:] - rdf_ref[1,:], c='red', lw=2)
plt.plot(reference_rdf100[0,:], reference_rdf100[1,:] - rdf_ref[1,:], label=f"100M L={round(loss, 5)}", c='blue', lw=1)

plt.legend(loc=2, title="Итоговый Loss")
plt.xlabel("r, Å")
plt.ylabel("Delta")
```
