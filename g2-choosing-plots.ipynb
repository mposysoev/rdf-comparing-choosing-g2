```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
etas = [0.125, 4, 4, 4, 4, 4, 4, 4]
r_shifts = [0, 3, 3.5, 4, 4.5, 5, 5.5, 6]

r = np.linspace(-1, 10, 500)

plt.figure(figsize=(16, 8))

total = np.array([0] * 500)
for i in range(len(etas)):
    g2_i = np.exp(-etas[i] * (r - r_shifts[i]) ** 2)
    total = total + g2_i
    plt.plot(
        r,
        g2_i,
        color=(155/255, 255/255, 1 / 6 * r_shifts[i]),
        label=f"eta = {etas[i]}; r_s = {r_shifts[i]}",
    )
    # plt.show()

# plt.plot(r, total / np.max(total), color="blue", label="total information")
plt.plot(r, total, color="blue", label="total information")
plt.legend()
```
