```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_ref = np.genfromtxt("water-CG-reference.rdf").T

data_training = np.genfromtxt("rdf_after_training0200.dat").T

data_add1 = np.genfromtxt("rdf_add_one0193.dat").T
```

```python
plt.figure(figsize=(16,8))

plt.plot(data_ref[0], data_ref[1], label="Reference", c="black")
plt.plot(data_training[0], data_training[1], label="After training", c="red")
plt.plot(data_add1[0], data_add1[1], label="After extend with 1 G2 at peak position = 2.8 A; eta = 7", c="green")

plt.legend()
```

```python
def calculate_loss_value(data, ref):
    ref = ref[1]
    data = data[1]

    s = np.sum(np.power(ref - data, 2))
    m = np.mean(np.power(ref - data, 2))

    print("Sum loss: ", s)
    print("MSE loss: ", m)
```

```python
calculate_loss_value(data_training, data_ref)
```

```python
calculate_loss_value(data_add1, data_ref)
```

```python
plt.figure(figsize=(16,8))

plt.plot(data_ref[0], data_ref[1], label="Reference", c="black")
plt.plot(data_training[0], data_training[1], label="After training", c="red")
plt.plot(data_add1[0], data_add1[1], label="After extend with 1 G2 at peak position = 2.8 A; eta = 7", c="green")

plt.legend()

plt.xlim(2, 6)
```
