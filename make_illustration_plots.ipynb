```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
```

```python
plt.rcParams.update({'font.size': 18})
```

```python
eps = 1.0
xs = np.linspace(-2.101, 2.507, 50000)
y = 0
func = xs**4 - 3.9 * xs**2 - xs + y**4 - 2 * y**2 + 1.3 * y - 0.4 * (xs + 2) ** 2
func = (func - np.min(func) + eps) 
func = func / np.max(func)


# Get the 'Blues' colormap
blues = cm.get_cmap('Blues')

# Retrieve the last color from the colormap
last_color = blues(0.8)  # 1.0 to get the last color

plt.figure(figsize=(10, 10))
plt.plot(xs, func, lw=5, color=last_color)
plt.xlabel("Neural Net Weight Space")
plt.ylabel("Loss Function Value")
plt.ylim(0, 1 + np.min(func))
plt.gca().set_xticks([]) 

plt.title("Illustration of Problem of Gradient Descending")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-2.0, 1.9, 500)
y = np.linspace(-1.7, 2.0, 500)

# Create meshgrid
X, Y = np.meshgrid(x, y)

# Define the function including Y influence more explicitly
Z = (X**4 - 3.9 * X**2 - X + Y**4 - 2 * Y**2 + 1.3 * Y - 0.4 * (X + 2) ** 2 - 0.4 * (Y - 2) ** 2)
Z = (Z - np.min(Z))
Z = Z / np.max(Z)

# Plotting
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surface = ax.plot_surface(X, Y, Z, cmap='Blues', edgecolor='none', alpha=0.8)

# Add a contour line at Y = 0
# First, find the index of y = 0 in the y array
y_zero_index = np.argmin(np.abs(y))
# Extract the Z values along y = 0
Z_at_y0 = Z[y_zero_index, :]

# Plot the line on the surface
ax.plot(x, np.zeros_like(x), Z_at_y0, color='r', linewidth=5, label="Y = 0")

# Enabling grid but hiding tick labels
ax.grid(True)  # Enable the grid

# To maintain grid lines but remove tick labels
ax.set_xticks(x[::50])  # Set ticks to be at regular intervals
ax.set_yticks(y[::50])  # Set ticks to be at regular intervals

# Hide the tick labels
ax.set_xticklabels([])  # Hide x-axis tick labels
ax.set_yticklabels([])  # Hide y-axis tick labels

# Labels and title
ax.set_xlabel("Neural Net Weight Space")
ax.set_ylabel("New Dimensions")
ax.set_zlabel("Loss Function Value", labelpad=6)
ax.set_title("New Ways in Higher Dimensions Spaces")

# Add legend
#ax.legend()

plt.show()
```
